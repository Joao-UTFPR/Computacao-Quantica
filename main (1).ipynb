```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_distribution

siml = AerSimulator()
qcirc = QuantumCircuit(2) # 2 quantum and 2 classical registers
qcirc.h(0) # applying Hadamard transform on qubit 0
qcirc.x(1) # applying Pauli X on qubit 1
qcirc.cx(0, 1) # applying CNOT on qubits 0 and 1 (controlled by qubit 0)
qcirc.measure_all()
qcirc.draw("mpl", initial_state=True) # MatPlotLib

```
