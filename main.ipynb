```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_distribution

siml = AerSimulator()
qcirc = QuantumCircuit(2, 2) # 2 quantum and 2 classical registers
qcirc.h(0) # applying Hadamard transform on qubit 0
qcirc.cx(0, 1) # applying CNOT on qubits 0 and 1 (controlled by qubit 0)
qcirc.measure(0, 0) # measure qubit 0 -> result on bit 0
qcirc.measure(1, 1) # measure qubit 1 -> result on bit 1
qcirc.draw("mpl")
```

```python
comp_qc = transpile(qcirc, siml) # compiled quantum circuit
job = siml.run(comp_qc, shots = 1000)  # run 1000 times
result = job.result()
counts = result.get_counts()   # result = dictionary
print("\nTotal count for 00 and 11 are:", counts)
```

```python
plot_histogram(counts)
```

```python
plot_distribution(counts)
```
